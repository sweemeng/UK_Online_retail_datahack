## Setup dependency

```python
import pandas
```

```python
from sklearn import cluster
```

```python
from matplotlib import pyplot
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

This is important, this is to make the dataviz work on notebook

```python
%matplotlib notebook
```

Now to load the customer data. 

```python
cust_df = pandas.read_excel("customers.xlsx")
```

```python
temp = cust_df[['CustomerID', 'Frequency', 'Recency', 'Monetary', 'FirstPurchase']]
```

Now we cluster. 

```python
centroid, labels, inertia = cluster.k_means(temp, n_clusters=5)
```

```python
temp["cluster"] = labels
```

Now we load the chart

```python
fig = pyplot.figure()
```

```python
ax = Axes3D(fig)
```

```python
ax.scatter(temp["Recency"], temp["Monetary"], temp["Frequency"], c=temp["cluster"])
```

```python
ax.set_xlabel("Recency")
```

```python
ax.set_ylabel("Monetary")
```

```python
ax.set_zlabel("Frequency")
```

```python
pyplot.show() # You don't actually need it on notebook, but useful for script
```
